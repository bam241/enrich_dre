```python
file = "case_2_/cascade_case_2_raw_S2.sqlite"


feed_case_2 = [] 
plt_feed_case_2 = []
for i in range(3):
    df_ = cyh.TransFrac(file, receivers[i], sender[i], nucs1, nucs2,uma_238u, uma_235u)
    df_ = pdh.RenameTS(pdh.RemoveNan(df_), "Time", c_names[i])
    df_['Time'] *= 1/12.
    feed_case_2.append(df_)
    plt_feed_case_2.append( (df_, c_names[i]) )
    
pdh.MakePlot(plt_feed_case_2, "Time", "Feed Enrichment" , loc="center right")
plt.axhline(y=0.9)
plt.xlim(0,8)
plt.ylim(0,1)




plt.savefig("assay_case_1_rec.png")
```
